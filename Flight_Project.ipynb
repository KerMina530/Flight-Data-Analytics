```python
#Q8 How does the price vary between classes 
#data.groupby('class')['price'].mean()
eco = data[data['class']== 'Economy']
eco.price.mean()
bus = data[data['class']== 'Business']
bus.price.mean()
mean_prices = [eco['price'].mean(), bus['price'].mean()]
classes = ['Economy', 'Business']
colors = ['skyblue', 'salmon']  # you can change these anytime
explode = [0, 0.1] 
plt.figure(figsize=(6,6))
plt.pie(
    mean_prices,
    labels=classes,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode=explode,
    shadow=True
)

plt.title('Average Ticket Price by Class')
plt.axis('equal')  
plt.show()
```

```python
#Separate Business vs Economy pricing, Detect outliers for pricing strategy
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10,6))
sns.boxplot(x='class', y='price', data=data)
plt.title("Ticket Price Distribution by Class")
plt.show()
```

```python
#More stops â†’ cheaper or more expensive?
plt.figure(figsize=(10,6))
sns.barplot(x='stops', y='price', data=data, estimator='mean')
plt.title("Average Ticket Price by Number of Stops")
plt.xlabel("Number of Stops")
plt.ylabel("Mean Price")
plt.show()
```

```python
#Longer flights cost more? confirming by correlation
plt.figure(figsize=(10,6))
sns.scatterplot(x='duration', y='price', data=data, hue='class')
plt.title("Flight Duration vs Ticket Price")
plt.show()

corr = data[['duration','price']].corr()
print("Correlation:\n", corr)
```

```python
#Which airline suffers more delays?
âœ” Delay effect on price?
if 'delay' in data.columns:
    plt.figure(figsize=(10,6))
    sns.barplot(x='airline', y='delay', data=data)
    plt.title("Average Delay by Airline")
    plt.xticks(rotation=45)
    plt.show()
else:
    print("No delay column found â€” skipping delay insights.")
```

```python
#Explore relationships including airlines, cabin class, cities
encoded_df = data.copy()
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
for col in encoded_df.select_dtypes(include=['object']).columns:
    encoded_df[col] = label.fit_transform(encoded_df[col])

plt.figure(figsize=(12,6))
sns.heatmap(encoded_df.corr(), annot=True, linewidths=0.5)
plt.title("Correlation Heatmap (Encoded Data)")
plt.show()
```

```python
#Interpret real numeric features (price, duration, stops, etc.)
numeric_df = data.select_dtypes(include=['int64', 'float64'])

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))
sns.heatmap(numeric_df.corr(), annot=True, linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```
